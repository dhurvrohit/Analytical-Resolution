```python
from tkinter import*
from tkinter import filedialog
from tkinter import messagebox
import numpy as np
import cv2
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk

win=Tk()
#win.geometry("1000x900")
win.state("zoomed")
win.resizable(height=False,width=False)
win.configure(bg='black')
win.title('RRS')

lbl_face=Label(win,text='Face Detection',font=('',60,'bold'),bg='black',fg='dark goldenrod').pack()
imageFrame=None
flag=False
BG=cv2.imread("C:/Users/ASUS/Desktop/bg.jpg")
cv2BG1=cv2.resize(BG,(1540,730))
cv2BG=cv2.cvtColor(cv2BG1,cv2.COLOR_BGR2RGB)
def log_out():
    option=messagebox.askyesno('confirmation','do u want to logout')
    if(option==True):
        home_screen()
    else:
        pass
def images_face_detection():
    
    def image_detect():
        frm=Frame(win)
        lmain=Label(frm,width=1535,height=730)
        lmain.grid(row=0,column=0)
        frm.place(relx=0,rely=.13,relwidth=1,relheight=1)
        img=Image.fromarray(cv2BG)
        imgtk=ImageTk.PhotoImage(img)
        lmain.imgtk=imgtk
        lmain.configure(image=imgtk)

        browse_button1=Button(frm,command=lambda:browse_image(),text='browse image',font=('',20,'bold'),bd=10).place(relx=.4,rely=.0)
        
        btn_back=Button(frm,command=lambda:welcome_screen(),text='BACK',font=('',20,'bold'),bd=10)
        btn_back.place(relx=.03,rely=.05)



    def browse_image():
        frm=Frame(win)
        lmain=Label(frm,width=1535,height=730)
        lmain.grid(row=0,column=0)
        frm.place(relx=0,rely=.13,relwidth=1,relheight=1)
        img=Image.fromarray(cv2BG)
        imgtk=ImageTk.PhotoImage(img)
        lmain.imgtk=imgtk
        lmain.configure(image=imgtk)

        global imageFrame
        if(imageFrame!=None):
            imageFrame.destroy()
        file_path=filedialog.askopenfilename()
        frame=cv2.imread(file_path)
        frame=cv2.resize(frame,(400,400))
        cv2image = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)

        imageFrame =Frame(win,width=700, height=500,bd=4,bg='green')
        lmain = Label(imageFrame)
        lmain.grid(row=0, column=0)
        imageFrame.place(relx=.37,rely=.3)
        img = Image.fromarray(cv2image)
        imgtk = ImageTk.PhotoImage(image=img)
        lmain.imgtk = imgtk
        lmain.configure(image=imgtk)
        button_detect=Button(frm,command=lambda:startface(frame,cv2image,lmain),text='detect face',font=('',20,'bold'),bd=10,width=12).place(relx=.4,rely=0)


    def startface(frame,cv2image,lmain):

        gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) 
        clf=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
        faces=clf.detectMultiScale(gray,1.2,2)
        for x,y,w,h in faces:
            cv2.rectangle(cv2image,(x,y),(x+w,y+h),(255,0,0),2)
        img = Image.fromarray(cv2image)
        imgtk = ImageTk.PhotoImage(image=img)
        lmain.imgtk = imgtk
        lmain.configure(image=imgtk)
        
    image_detect()
    
def webcam_face_detection():
    
    def webcam():
        frm=Frame(win)
        lmain=Label(frm,width=1535,height=730)
        lmain.grid(row=0,column=0)
        frm.place(relx=0,rely=.13,relwidth=1,relheight=1)
        img=Image.fromarray(cv2BG)
        imgtk=ImageTk.PhotoImage(img)
        lmain.imgtk=imgtk
        lmain.configure(image=imgtk)
        
        button_start=Button(frm,command=lambda:start(),text='start camera',font=('',20,'bold'),bd=10).place(relx=.01,rely=.2)
        button_stop=Button(frm,command=lambda:stop(),text='stop camera',font=('',20,'bold'),bd=10).place(relx=.01,rely=.4)
           
        btn_logout=Button(frm,command=lambda:welcome_screen(),text="BACK",font=('',20,'bold'),bd=10)
        btn_logout.place(height = 50,relx=.85,rely=0) 
    def show_frame():
        _, frame = cap.read()
        cv2image = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) 
        clf=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
        faces=clf.detectMultiScale(gray,1.3,7)
        for x,y,w,h in faces:
            if(flag==True):
                cv2.rectangle(cv2image,(x,y),(x+w,y+h),(255,0,0),2)
        img = Image.fromarray(cv2image)
        imgtk = ImageTk.PhotoImage(image=img)
        lmain.imgtk = imgtk
        lmain.configure(image=imgtk)
        lmain.after(10, show_frame) #calls show_frame after 10 mills
    
    def start():
        global lmain,cap,lmain,imageFrame
        cap = cv2.VideoCapture(0)
        imageFrame = tk.Frame(win,width=700, height=500,bd=10,bg='green')
        lmain = tk.Label(imageFrame)
        lmain.grid(row=0, column=0)
        imageFrame.place(relx=.25,rely=.15)
        button_start_detect=Button(imageFrame,command=lambda:startface(),text='detect face',font=('',20,'bold'),bd=10,width=12).place(relx=.1,rely=.9)
        button_stop_detect=Button(imageFrame,command=lambda:stopface(),text='stop detecting',font=('',20,'bold'),bd=10,width=12).place(relx=.5,rely=.9)
        
        show_frame()

    def stop():
        cap.release()
        lmain.destroy()
        imageFrame.destroy()

    def startface():
        global flag
        flag=True

    def stopface():
        global flag
        flag=False
    webcam()
    
def videos_face_detection():
    
    def video_face():
        frm=Frame(win)
        lmain=Label(frm,width=1535,height=730)
        lmain.grid(row=0,column=0)
        frm.place(relx=0,rely=.13,relwidth=1,relheight=1)
        img=Image.fromarray(cv2BG)
        imgtk=ImageTk.PhotoImage(img)
        lmain.imgtk=imgtk
        lmain.configure(image=imgtk)
        
        entry_user = Entry(frm,font=('',20,'bold'),bd=10)
        entry_user.place(height = 50,width = 300 ,relx=.38,rely=.1)
        entry_user.focus()
    
        btn_logout=Button(frm,command=lambda:welcome_screen(),text="BACK",font=('',20,'bold'),bd=10)
        btn_logout.place(height = 50,relx=.85,rely=0) 
        
        browse_button1=Button(frm,command=lambda:browse_video(entry_user),text='browse video',font=('',20,'bold'),bd=10).place(relx=.4,rely=.0)
        
            
        btn_detect = Button(frm,command=lambda:video_det(entry_user),text='Detect',font=('',20,'bold'),bd=10)
        btn_detect.place(relx=.43,rely=.2)
    
    def browse_video(src):
        file_path = filedialog.askopenfilename()
        src.delete(0,END)
        src.insert(0,file_path)
    
    def video_det(entry_user):      
       
        src_path = entry_user.get()
        if(len(src_path)==0):
            messagebox.showerror("error","Please select video")
            video_face()
        else:      
            vdo=cv2.VideoCapture(src_path)
            clf=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
            while(True):
                flag,img=vdo.read()
                if(flag==True):
                    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
                    faces=clf.detectMultiScale(gray,1.4,4)
                    for x,y,w,h in faces:
                        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
                    cv2.imshow('vdo',img)
                    k=cv2.waitKey(1)
                    if(k==ord('q')):
                        video_face()
                        break
                else:
                    break
        vdo.release()
        cv2.destroyAllWindows()
        
    video_face()
    
def home_screen():
    frm=Frame(win,bg='gold')
    lmain=Label(frm,width=1535,height=730)
    lmain.grid(row=0,column=0)
    frm.place(relx=0,rely=.13,relwidth=1,relheight=1)
    img=Image.fromarray(cv2BG)
    imgtk=ImageTk.PhotoImage(img)
    lmain.imgtk=imgtk
    lmain.configure(image=imgtk)

    lbl_username=Label(frm,text='USERNAME:',font=('',20,'bold'),bg='black',fg='dark goldenrod')
    lbl_username.place(relx=.3,rely=.2)

    entry_use=Entry(frm,font=('',20,'bold'),bd=10)
    entry_use.place(relx=.43,rely=.2)
    entry_use.focus()

    lbl_password=Label(frm,text='PASSWORD:',font=('',20,'bold'),bg='black',fg='dark goldenrod')
    lbl_password.place(relx=.3,rely=.35)

    entry_pass=Entry(frm,font=('',20,'bold'),bd=10,show='*')
    entry_pass.place(relx=.43,rely=.35)

    btn_login=Button(frm,command=lambda:authentication(entry_use,entry_pass),text='login',font=('',20,'bold'),bd=10,width=5)
    btn_login.place(relx=.5,rely=.5)
    
def authentication(user,pwd):
    user=user.get()
    pwd=pwd.get()
    if(len(user)==0 or len(pwd)==0):
        messagebox.showwarning("validation","Please fill both fields")
        return
    elif(user!='RRS' or pwd!='RRS'):
        messagebox.showerror("fail",'invalid username/password')
    else:    
        if(user=='RRS' and pwd=='RRS'):
            welcome_screen()
            
def welcome_screen():
        
        frm=Frame(win)
        lmain=Label(frm,width=1535,height=730)
        lmain.grid(row=0,column=0)
        frm.place(relx=0,rely=.13,relwidth=1,relheight=1)
        img=Image.fromarray(cv2BG)
        imgtk=ImageTk.PhotoImage(img)
        lmain.imgtk=imgtk
        lmain.configure(image=imgtk)

        btn_Single_feed=Button(frm,text='using image',command=lambda:images_face_detection(),font=('',20,'bold'),bd=10,width=15)
        btn_Single_feed.place(relx=.4,rely=.2)

        btn_bulk_feed=Button(frm,text='using video',command=lambda:videos_face_detection(),font=('',20,'bold'),bd=10,width=15)
        btn_bulk_feed.place(relx=.4,rely=.4)
        
        btn_bulk_feed=Button(frm,text='using webcam',command=lambda:webcam_face_detection(),font=('',20,'bold'),bd=10,width=15)
        btn_bulk_feed.place(relx=.4,rely=.6)

        btn_log_out=Button(frm,text='LOG OUT',command=lambda:log_out(),font=('',20,'bold'),bd=5,)
        btn_log_out.place(relx=.8,rely=0)

    
home_screen()
win.mainloop()
```
